```python
import vizdoom as vzd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
from mazeexplorer import MazeExplorer
train_env = MazeExplorer(number_maps=1,
                         size=(15, 15),
                         random_spawn=True,
                         random_textures=False,
                         keys=6)
```

```python
game = vzd.DoomGame()
game.load_config('resources/temp_maps/datagen.cfg')
game.set_doom_scenario_path('resources/temp_maps/1.wad')
game.set_automap_mode(vzd.AutomapMode.WHOLE)
game.set_objects_info_enabled(True)
game.set_labels_buffer_enabled(True)
game.init()
game.new_episode()
game.make_action([1,0,0,0])
state = game.get_state()
am = state.automap_buffer.transpose(1,2,0)
obj = state.objects
labels = state.labels
game.close()
print(state.game_variables)
plt.imshow(am)
```
